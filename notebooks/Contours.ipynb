```julia
canvases = []
for contour in contours
    grid =zeros(Int, 10, 10)
    for ind in contour
        grid[ind] = 1
    end
    push!(canvases, plot(Gray.(grid)))
end

plot(canvases..., layout=(1, length(canvases)), border=:none, size=(800, 200))
```

```julia
canvases = []
for polygon in contours
    inside = []
    points = []
    for x in 1:10
        for y in 1:10
            push!(inside, point_in_polygon(CartesianIndex(x, y), polygon))
            push!(points, (y, x))
        end
    end
    
    x = [point[1] for point in points]
    y = [point[2] for point in points]
    color = [is_inside ? "red" : "blue" for is_inside in inside]
    canvas = scatter(x, y, color=color, label="")
    x = [point[2] for point in polygon]
    y = [point[1] for point in polygon]
    plot!(canvas, vcat(x, x[1]), vcat(y, y[1]), aspectratio=1, label="", color=:yellow, ylims=(0, 12))
    push!(canvases, canvas)
end
plot(canvases..., layout = (1, length(contours)), size=(800, 200), yflip=true)
```
